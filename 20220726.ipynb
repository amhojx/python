# request 使用簡介

```python
import requests
url = "https://victorgau.github.io/khpy_web_intro/html_only.html"
requests.get(url).text
```

```python
url = "https://tcgbusfs.blob.core.windows.net/dotapp/youbike/v2/youbike_immediate.json"
data = requests.get(url).json()
type(data)
```

```python
data[0]
```

```python
for i in range(100):
  print(data[i]['sna'],":(",data[i]['lat'],",",data[i]['lng'],")")
```

```python
gps = [data[0]['lat'], data[0]['lng']]
map = folium.Map(location = gps, zoom_start=13)
for i in range(100):
  # print(data[i]['sna'],":(",data[i]['lat'],",",data[i]['lng'],")")
  folium.Marker((data[i]['lat'], data[i]['lng']), popup = data[i]['sna']).add_to(map)
map
```

```python
gps = [data[0]['lat'], data[0]['lng']]
map = folium.Map(location = gps, zoom_start=13)
for x in data:
  # print(data[i]['sna'],":(",data[i]['lat'],",",data[i]['lng'],")")
  folium.Marker((x['lat'], x['lng']), popup = x['sna']).add_to(map)
map
map.save("index.html")
```
